```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data into a pandas DataFrame
df = pd.read_csv('/Users/dylanfunk/my_project/NFL_Data.csv')

# Display the first few rows of the data
print("First few rows of the data:")
print(df.head())

# Basic Data Information
print("\nBasic Info:")
print(df.info())

# Summary statistics
print("\nSummary statistics:")
print(df.describe())

# Data cleaning - Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Data Analysis 1: Top 5 players by Total Yards
top_5_players = df[['Player Name', 'Total Yards']].sort_values(by='Total Yards', ascending=False).head(5)
print("\nTop 5 players by Total Yards:")
print(top_5_players)

# Data Analysis 2: Players with the most touchdowns
top_touchdowns = df[['Player Name', 'Touchdowns']].sort_values(by='Touchdowns', ascending=False).head(5)
print("\nTop players by Touchdowns:")
print(top_touchdowns)

# Data Analysis 3: Distribution of Total Yards
plt.figure(figsize=(10, 6))
sns.histplot(df['Total Yards'], kde=True, color='blue')
plt.title('Distribution of Total Yards')
plt.xlabel('Total Yards')
plt.ylabel('Frequency')
plt.show()

# Data Analysis 4: Players' Performance Visualization (Total Yards vs Touchdowns)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Total Yards', y='Touchdowns', hue='Position', palette='coolwarm')
plt.title('Total Yards vs Touchdowns by Position')
plt.xlabel('Total Yards')
plt.ylabel('Touchdowns')
plt.legend(title='Position')
plt.show()

# Data Analysis 5: Injury Status Count
injury_status_count = df['Injury Status'].value_counts()
print("\nInjury Status Count:")
print(injury_status_count)

# Plotting Injury Status Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Injury Status', palette='Set2')
plt.title('Distribution of Injury Status')
plt.xlabel('Injury Status')
plt.ylabel('Count')
plt.show()

# Data Analysis 6: Pass Completion Percentage Distribution (for QBs)
qbs = df[df['Position'] == 'QB']
plt.figure(figsize=(10, 6))
sns.histplot(qbs['Pass Completion Percentage'], kde=True, color='green')
plt.title('Pass Completion Percentage Distribution for QBs')
plt.xlabel('Pass Completion Percentage')
plt.ylabel('Frequency')
plt.show()

# Data Analysis 7: Correlation between Total Yards and Touchdowns
correlation = df[['Total Yards', 'Touchdowns']].corr()
print("\nCorrelation between Total Yards and Touchdowns:")
print(correlation)

# Visualization of Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=2)
plt.title('Correlation Heatmap: Total Yards vs Touchdowns')
plt.show()
```
